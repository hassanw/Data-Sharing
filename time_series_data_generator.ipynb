```python
import timesynth as ts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (12,9)
```

```python
time_sampler = ts.TimeSampler(stop_time=100)
regular_time_samples = time_sampler.sample_regular_time(num_points=20000)
sinusoid = ts.signals.Sinusoidal(frequency=0.25)
timeseries = ts.TimeSeries(sinusoid)
samples, signals, errors = timeseries.sample(regular_time_samples)
plt.plot(regular_time_samples, samples, marker='o', markersize=4)
```

```python
df = pd.concat([pd.Series(regular_time_samples,name="time"),pd.Series(samples,name="value")],axis=1)
#df.to_csv("sine_wave_timeseries_data.csv")
```
